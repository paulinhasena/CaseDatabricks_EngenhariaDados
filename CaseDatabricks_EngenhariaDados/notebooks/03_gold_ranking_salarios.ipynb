```python
"""
Ranking de Profissões por Salário
Objetivo: Gerar o ranking de cargos com base no maior salário com dados vindos da silver.

Observação:
Este notebook depende das tabelas SILVER atualizadas.
Reexecução necessária apenas quando houver mudanças na camada SILVER.
"""
```

```python
from pyspark.sql import functions as F
```

```python
df_rem = spark.table("silver_remuneracao")
df_cad = spark.table("silver_cadastro")
```

```python
df_base = (
    df_rem.join(
        df_cad,
        on="id_servidor",
        how="inner"
    )
)
```

```python
df_base.count()
```

```python
df_ranking_final = (
    df_base
        .groupBy("cargo")
        .agg(
            F.max("vlr_remuneracao_bruta_brl").alias("maior_salario_brl"),
            F.max("vlr_remuneracao_bruta_usd").alias("maior_salario_usd")
        )
        .filter(F.col("maior_salario_brl") > 0)
        .orderBy(F.col("maior_salario_brl").desc())
)
```

```python
df_ranking_final.filter(F.col("maior_salario_brl") <= 0).count()
```

```python
df_ranking_final.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_ranking_profissoes_salario")
```
